```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 22, 28],
    'Score': [85, 92, 78, 90, 88],
    'City': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Los Angeles']
}

df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
```

```python
# --- Data Analysis
```

```python
# --- Descriptive Statistics
```

```python
print("\nDescriptive Statistics:")
print(df.describe())
```

```python
# --- Grouping Data
```

```python
grouped_data = df.groupby('City').mean()
print("\nAverage Score by City:")
print(grouped_data)
```

```python
# --- Correlation Matrix
```

```python
correlation_matrix = df.corr()
print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# --- Data Visualization
```

```python
# --- Line Plot
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['Name'], df['Score'], marker='o')
plt.title('Scores by Name')
plt.xlabel('Name')
plt.ylabel('Score')
plt.grid()
plt.show()
```

```python
# --- Bar Plot
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='Name', y='Score', data=df)
plt.title('Scores by Name')
plt.show()
```

```python
# --- Histogram
```

```python
plt.figure(figsize=(10, 5))
plt.hist(df['Score'], bins=5, color='skyblue', edgecolor='black')
plt.title('Distribution of Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# --- Box Plot
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='City', y='Score', data=df)
plt.title('Score Distribution by City')
plt.show()
```

```python
# --- Heatmap
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# --- Customizing Plots
```

```python
data = {'Name': ['Alice', 'Bob', 'Charlie'],
        'Score': [90, 85, 88]}
df = pd.DataFrame(data) 

plt.figure(figsize=(10, 5))
sns.set(style="whitegrid")
sns.barplot(x='Name', y='Score', data=df, hue='Name', palette='Blues', legend=False)
plt.title('Scores by Name', fontsize=16)
plt.xlabel('Name', fontsize=12)
plt.ylabel('Score', fontsize=12)
plt.show()
```
